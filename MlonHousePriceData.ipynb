# Encoding

```python
#step 1 : Divied the Data based on their datatype
df_cat = df_house_price.select_dtypes(include=['object'])
df_cat.columns
```

```python
# use dummmies coding
dummy_var = pd.get_dummies(data= df_cat, drop_first =True)
dummy_var =dummy_var.reset_index()
```

```python
dummy_var2 = pd.get_dummies(data= df_cat)
```

```python
dummy_var.head(2)
```

```python
df_num =df_house_price.select_dtypes(include= [np.number])
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#store the target variable 'chance of admit in a dataframe
df_target = df_house_price['Price']

df_num =df_num.drop('Price',axis=1)
```

```python
# Initiate the standardscaler
ss = StandardScaler()
```

```python
scale = ss.fit_transform(df_num)
scale
```

```python
df_scale = pd.DataFrame(scale , columns = df_num.columns)
df_scale
```

```python
# make a dataframe for independent variable 
df_independence = pd.concat([df_scale,dummy_var], axis=1)
df_independence.head()
```

# Splits the data for training and testing

```python
import statsmodels
import statsmodels.api as sm

#add a concat columns to the dataframe
df_independence = sm.add_constant(df_independence)
df_independence
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(df_independence, df_target, random_state = 10, test_size=0.2)
```

```python
#print  dimension of train set
print('X_train',X_train.shape)
print('Y_train',Y_train.shape)

#print  dimension of train set
print('X_test',X_test.shape)
print('Y_test',Y_test.shape)

```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# create a linear regression model
model =LinearRegression()
```

```python
#fit the model
model.fit(X_train , Y_train)
```

```python
# predict on test data
prediction = model.predict(X_test)
```

```python
# predict on test data
mse = mean_squared_error(Y_test, prediction)
mse
```
